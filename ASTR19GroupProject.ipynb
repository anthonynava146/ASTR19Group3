```python
import numpy as np
```

```python
data = np.genfromtxt("ASTR19_F24_group_project_data.txt", dtype = [('day', 'i8'), ('time', 'U6'), ('height', 'f8')])

day = []
time = []
height = []

#Get data from file
for i in range(82):
    day.append(data[i][0])
    time.append(data[i][1])
    height.append(data[i][2])
    
print(day)
print(' ')
print(time)
print(' ')
print(height)
```

```python
from datetime import datetime

total_array = []

for i in range(82):
    hours,  minutes = map(int, time[i].split(":"))
    total_min = hours * 60 + minutes
    total_time = total_min / (24 * 60)
    
    total_array.append(total_time)
    
time_array = np.array(total_array)
time_array
```

```python
x_axis = time_array + day
x_axis
```

```python
import matplotlib.pyplot as plt

#Plot our time of day vs tide height data
plt.plot(x_axis, height, color='#1f77b4', label='Tide')
plt.title('Tide Height vs Time of Day')
plt.xlabel('Time (Days from Start)')
plt.ylabel('Tide Height (feet)')
plt.legend()
plt.show()
```

```python
from scipy import optimize

#Oscillation function for the line of best fit
def tidal_oscillation(x, A, B, C, D, E, F, G, H):
    return (A * np.sin(B * x + C) + D) * E * np.sin(F * x + G) + H

#Fitting the line to the data
params, covariance = optimize.curve_fit(tidal_oscillation, x_axis, height, p0=[1, 2* np.pi/12.48, 1, 1, 1, 2*np.pi, 1, 1], sigma=np.full(len(height), 0.25))

#Parameters for oscillation function
a_fit, b_fit, c_fit, d_fit, e_fit, f_fit, g_fit, h_fit = params

#Establishing the x and y axes for the fitted line
x_fit = np.linspace(np.min(x_axis), np.max(x_axis), 1000)
tide_fit = tidal_oscillation(x_fit, a_fit, b_fit, c_fit, d_fit, e_fit, f_fit, g_fit, h_fit)

# Plotting
plt.figure(figsize=(12, 6))
plt.scatter(x_axis, height, color='blue', label='Observed Tide Heights', alpha=0.6)
plt.plot(x_fit, tide_fit, color='red', label='Fitted Tide Model')
plt.title('Fitted Tide Height Model vs. Observed Data')
plt.xlabel('Time (days from start)')
plt.ylabel('Tide Height (feet)')
plt.legend()
plt.savefig("modelfit.pdf", bbox_inches='tight', facecolor='white')
```

```python
#Establishing data for calculating residuals
y_data = np.array(height)
y_pred = tidal_oscillation(x_axis, *params)

#Calculating residuals
residuals = y_data - y_pred

#Printing the residuals, the mean, and the STD of the residuals
print(f"Residuals: {residuals}\n")

res_mean = np.mean(residuals)
print(f"Residuals Mean: {res_mean}")

res_std = np.std(residuals)
print(f"Residuals STD: {res_std}")

```

```python
#Create histogram
bins = 30
plt.figure(figsize=(8, 5))
plt.hist(residuals, bins, alpha=0.5, edgecolor='black', color='#1f77b4', label='Residuals')
plt.title('Histogram of Residuals')
plt.xlabel('Residuals of 20')
plt.ylabel('Magnitude')
plt.axvline(res_mean,color='black', linestyle='dashed', label='Mean')
plt.axvline(res_mean + res_std, color='red',alpha=0.6, label='Mean ± STD')
plt.axvline(res_mean - res_std, color='red', alpha=0.6)
plt.axvspan(res_mean,res_std, color='red',alpha=0.1, label='± STD')
plt.axvspan(res_mean,-res_std, color='red',alpha=0.1)
plt.savefig('residual_histogram.pdf', bbox_inches='tight', facecolor='white')
plt.legend()
plt.show()

#Calculate intrinsic scatter
int_scatter = np.sqrt((np.square(res_std))-(np.square(0.25)))
print(f"Intrinsic Scatter: {int_scatter}")
```

```python
#2ft deviation from the tsunami
tsunamiDev = 2.0

deviationSTD = tsunamiDev / res_std

#Change height of high tide by 2ft on January 14th
height[26] = height[26] + tsunamiDev

#Recalculate residuals 
y_data = np.array(height)
y_pred = tidal_oscillation(x_axis, *params)
new_residuals = y_data - y_pred
print(f"Residuals: {new_residuals}")

#Recalucalte residual mean and std for the histogram
new_res_mean = np.mean(new_residuals)
new_res_std = np.std(new_residuals)

#Recalculate intrinsic scatter for the new data
int_scatter = np.sqrt((np.square(new_res_std))-(np.square(0.25)))

#Create new histgram to account for the tsunami
plt.figure(figsize=(8,5))
plt.hist(new_residuals, bins, edgecolor='black', alpha=0.6, label='Tsunami Res')
plt.title("Histogram of Residuals with 2ft Tsunami Outlier")
plt.xlabel("Residuals (Feet)")
plt.ylabel("Frequency")
plt.plot(new_res_std)
plt.axvline(new_res_mean,color='black', linestyle='dashed', label='Mean')
plt.axvline(new_res_mean + new_res_std, color='red',alpha=0.6, label='Mean ± STD')
plt.axvline(new_res_mean - new_res_std, color='red', alpha=0.6)
plt.axvspan(new_res_mean, new_res_std, color='red',alpha=0.1, label='± STD')
plt.axvspan(new_res_mean, -new_res_std, color='red',alpha=0.1)
plt.savefig('tsunami_residual_histogram.pdf', bbox_inches='tight', facecolor='white')
plt.legend()
plt.show()

#Print data
print(f"Standard deviation of new residuals: {new_res_std}")
print(f"Intrinsic Scatter: {int_scatter}")
print(f"The 2ft tsunami deviation corresponds to {deviationSTD:.8f} standard deviations.")
```
