```python
import numpy as np

data = np.genfromtxt("ASTR19_F24_group_project_data.txt", dtype=[('myint','i8'),('mystring','S5'),('myfloat', 'f8')])

day = []
time = []
height = []

for i in range(82):
    day.append(data[i][0])
    time.append(data[i][1].decode("utf-8"))
    height.append(data[i][2])
    
print(day)
print(' ')
print(time)
print(' ')
print(height)
```

```python
#Use line fitting between both y values
#Two oscillations, one immediate frequency, one general frequency
#ONE CURVE
from datetime import datetime

time_in_days = [datetime.strptime(i, '%H:%M').hour / 24 + datetime.strptime(i, '%H:%M').minute / (24 * 60) for i in time]
time_array = np.array(time_in_days)
time_array
```

```python
x_axis = time_array + day
x_axis
```

```python
import matplotlib.pyplot as plt
plt.plot(x_axis, height)
```

```python
# a = 10
# b = 0.1
# c = 0.27
# d = -1.3
# sig = 0.25

from scipy import optimize

def f_line(x,a,b,c,d):
    return a * np.sin(b * x_axis + c) + d
params, params_cov = optimize.curve_fit(f_line, x_axis, height, p0= [1, 2 * np.pi, 0.1, -0.1])

a_fit= params[0]
b_fit= params[1]
c_fit= params[2]
d_fit= params[3]

x_fit = np.linspace(0,np.max(x_axis), 1000)
y_fit = a_fit * np.sin(b_fit * x_fit + c_fit) + d_fit

plt.plot(x_fit, y_fit)
```

```python
from scipy import optimize

def tidal_oscillation(x, A, B, C, D, E, F, G, H):
    return (A * np.sin(B * x + C) + D) * E * np.sin(F * x + G) + H

y_mean = np.mean(height)

params, covariance = optimize.curve_fit(tidal_oscillation, x_axis, height, p0=[1, 2* np.pi/12.48, 1, 1, 1, 2*np.pi, 1, 1], sigma=np.full(len(height), 0.25))
a_fit, b_fit, c_fit, d_fit, e_fit, f_fit, g_fit, h_fit = params               #A,        B,       C,  D,  E,    F,    G, H

x_fit = np.linspace(np.min(x_axis), np.max(x_axis), 1000)
tide_fit = tidal_oscillation(x_fit, a_fit, b_fit, c_fit, d_fit, e_fit, f_fit, g_fit, h_fit)



# Plotting
plt.figure(figsize=(12, 6))
plt.scatter(x_axis, height, color='blue', label='Observed Tide Heights', alpha=0.6)
plt.plot(x_fit, tide_fit, color='red', label='Fitted Tide Model')
plt.title('Fitted Tide Height Model vs. Observed Data')
plt.xlabel('Time (days from start)')
plt.ylabel('Tide Height (feet)')
plt.legend()
plt.savefig("modelfit.pdf", bbox_inches='tight', facecolor='white')
```

```python
y_data = np.array(height)
y_pred = tidal_oscillation(x_axis, * params)

residuals = y_data - y_pred
print(f'Residuals: {residuals}')

res_mean = np.mean(residuals)
print(f'Residuals Mean: {res_mean}')

res_std = np.std(residuals)
print(f'Residuals Standard Deviation:{res_std} ')
```
