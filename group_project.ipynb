```python
import numpy as np

data = np.genfromtxt("ASTR19_F24_group_project_data.txt", dtype=[('myint','i8'),('mystring','S5'),('myfloat', 'f8')])

day = []
time = []
height = []

for i in range(82):
    day.append(data[i][0])
    time.append(data[i][1].decode("utf-8"))
    height.append(data[i][2])
    
print(day)
print(' ')
print(time)
print(' ')
print(height)
```

```python
import matplotlib.pyplot as plt

plt.scatter(day, height)
plt.xlabel("day")
plt.ylabel("height in inches")

#notice how every x value has 2 y values, there are 2 curves superimposed on eachother
#How do we turn those into just a single curve on x?
```
